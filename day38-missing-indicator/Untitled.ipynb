```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split, GridSearchCV

from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.linear_model import LogisticRegression
```

```python
df = pd.read_csv('train.csv')
df.head(5)
```

```python
df.drop(columns=['PassengerId','Name','Ticket','Cabin'],inplace=True)
```

```python
imputer_scalling_num = Pipeline(
    steps=[
        ('num_imput',SimpleImputer(strategy='median')),
        ('num_scalling',StandardScaler()),
        
    ]

)


imputer_scalling_cat = Pipeline(
    steps=[
        ('cat_imput',SimpleImputer(strategy='mosr_frequent')),    
        ('cat_scalling',OneHotEncoder(sparse=True))    
        
    ]

)
```

```python
numerical_features = ['Age', 'Fare']
categorical_features = ['Embarked', 'Sex']
trans = ColumnTransformer(
    transformers=[
        ('num',imputer_scalling_num, numerical_features),
        ('cat', imputer_scalling_cat, categorical_features)
    ]
)
```

```python
categorical_features = ['Embarked', 'Sex']
```

```python
clf = Pipeline(steps=[
    ('preprocessor', trans),
    ('classifier', LogisticRegression())
])
```
